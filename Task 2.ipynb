Create the Decision Tree classifier and visualize it graphically. 

```python
# Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.datasets import load_iris
```

```python
li = load_iris()

# Reading the data By converting Normal String DataSet to Raw String DataSet

irisdata = pd.read_csv(r"C:\Users\satyaki\Downloads\IRIS.csv")
```

```python
#Printing First 5 Rows

irisdata.head()
```

```python
#Printing Last 5 Rows

irisdata.tail()
```

```python
data = irisdata.values
X=data[:, 0:4]
Y=data[:,4]
X[0:4]
```

```python
Y[0:4]
```

## Splitting Data

```python
# Split dataset into training set and test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1) # 80% training and 20% test
```

## Building Decision Tree Model

```python
# Create Decision Tree classifer object
tree = DecisionTreeClassifier()

# Train Decision Tree Classifer
tree.fit(X_train,Y_train)
```

```python
#Predict the response for test dataset
Yprediction = tree.predict(X_test)
```

### Evaluating Model

```python
# Model Accuracy, how often is the classifier correct?

print("Accuracy Of Decision Tree:",metrics.accuracy_score(Y_test, Yprediction))
```

### Visualizing Decision Trees

```python
from six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus


dot_data= StringIO()
export_graphviz(tree, out_file=dot_data, filled=True, rounded=True, special_characters=True, feature_names = li.feature_names,class_names=li.target_names)
g = pydotplus.graph_from_dot_data(dot_data.getvalue())  
g.write_png('species.png')
Image(g.create_png())
```

## Optimizing Decision Tree Performance

```python
# Create Decision Tree classifer object
tree = DecisionTreeClassifier(criterion="entropy", max_depth=3)

# Train Decision Tree Classifer
tree = tree.fit(X_train,Y_train)

#Predict the response for test dataset
Yprediction = tree.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(Y_test, Yprediction))
```
